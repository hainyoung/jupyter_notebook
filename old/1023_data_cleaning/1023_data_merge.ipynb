```python
import pandas as pd
import numpy as np
```

```python
log_df = pd.read_csv("./1023_gps_log_edit.csv", encoding = "utf-8")
out_df = pd.read_csv("./1023_raw_output.csv", encoding = "utf-8")

inter_df = pd.read_csv("./1023_data_cleaning/1023_inter_data.csv", encoding = "utf-8")
ffill_df = pd.read_csv("./1023_data_cleaning/1023_ffill_data.csv", encoding = "utf-8")
```

```python
# df = pd.merge(log_df, out_df, on = "time", how = "outer")
df1 = pd.merge(log_df, inter_df, on = "time")
df1 = df1.rename({'date_x':'date'}, axis = 'columns')
df1 = df1[['date', 'time', 'provider', 'device_name', 'latitude', 'longtitude', 'speed', 'ldws', 'lkas']]
df1
```

```python
df1.isnull().sum()
```

```python
df2 = pd.merge(log_df, inter_df, on="time", how="right")
df2 = df2.rename({'date_y':'date'}, axis = 'columns')
df2 = df2[['date', 'time', 'provider', 'device_name', 'latitude', 'longtitude', 'speed', 'ldws', 'lkas']]
df2
```

```python
df3 = pd.merge(log_df, out_df, on = "time")
df3 = df3.rename({'date_x':'date'}, axis = 'columns')
df3 = df3[['date', 'time', 'provider', 'device_name', 'latitude', 'longtitude', 'speed', 'ldws', 'lkas']]
df3
```

```python
df3.isnull().sum()
```

```python
df4 = pd.merge(log_df, out_df, on = "time", how = "right")
df4 = df4.rename({'date_y':'date'}, axis = 'columns')
df4 = df4[['date', 'time', 'provider', 'device_name', 'latitude', 'longtitude', 'speed', 'ldws', 'lkas']]
df4
```

```python
df4.isnull().sum()
```

```python
df1.to_csv("./1023_inner_merge_interpolate.csv", index = False)
df2.to_csv("./1023_right_merge_interpolate.csv", index = False)
df3.to_csv("./1023_inner_merge_rawdata.csv", index = False)
df4.to_csv("./1023_right_merge_rawdata.csv", index = False)
```
