```python
import pandas as pd
import numpy as np
```

```python
distance_data_raw = pd.read_csv('./1107_distance_raw_data.csv', encoding = 'utf-8')
distance_data_raw
```

```python
distance_data_raw['rtime'] = distance_data_raw['date'] + " " + distance_data_raw['time']
distance_data_raw
```

```python
distance_data_raw = distance_data_raw[['rtime', 'distance']]
distance_data_raw
```

```python
distance_data_raw.isnull().sum()
```

```python
distance_data_clean = distance_data_raw.dropna(axis=0)
distance_data_clean
```

```python
distance_data_clean.isnull().sum()
```

```python
# distance_data_clean = distance_data_clean
distance_data_clean.duplicated(['rtime']).sum()
```

```python
distance_clean_data = distance_data_clean.drop_duplicates('rtime', keep='first')
```

```python
distance_clean_data.duplicated(['rtime']).sum()
```

```python
distance_clean_data
```

```python
distance_clean_data.isnull().sum()
```
